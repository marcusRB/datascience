```python
# Create 4 new variables, dayofweek number and name, month of day and hour of day.
#cabify2010['start_at'] = pd.to_datetime(cabify2010['start_at'], format="%m/%d/%Y %H:%M:%S")
cabify2010['DayOfWeekNum'] = cabify2010['start_at'].dt.dayofweek
cabify2010['DayOfWeek'] = cabify2010['start_at'].dt.weekday_name
cabify2010['MonthDayNum'] = cabify2010['start_at'].dt.day
cabify2010['HourOfDay'] = cabify2010['start_at'].dt.hour
```

### Alternative datetime

```python
#converting start_at column to datetime 
#cabify2010['start_at'] =  pd.to_datetime(cabify2010['start_at'], format='%Y-%m-%d %H:%M:%S')


#extracting the Hour from start_at column and putting it into a new column named 'Hour'
cabify2010['Hour'] = cabify2010.start_at.apply(lambda x: x.hour)

#extracting the Minute from start_at column and putting it into a new column named 'Minute'
cabify2010['Minute'] = cabify2010.start_at.apply(lambda x: x.minute)

#extracting the Month from start_at column and putting it into a new column named 'Month'
cabify2010['Month'] = cabify2010.start_at.apply(lambda x: x.month)

#extracting the Day from start_at column and putting it into a new column named 'Day'
cabify2010['Day'] = cabify2010.start_at.apply(lambda x: x.day)

#extracting the Weekday from start_at column and putting it into a new column named 'WeekDay'
cabify2010['WeekDay'] = cabify2010.start_at.apply(lambda x: x.strftime('%A'))
```

```python
tableau_color_blind = [(0, 107, 164), (255, 128, 14), (171, 171, 171), (89, 89, 89),
             (95, 158, 209), (200, 82, 0), (137, 137, 137), (163, 200, 236),
             (255, 188, 121), (207, 207, 207)]

for i in range(len(tableau_color_blind)):  
    r, g, b = tableau_color_blind[i]  
    tableau_color_blind[i] = (r / 255., g / 255., b / 255.)
```

```python
sns.set_style('whitegrid')

## set palette   
current_palette = sns.color_palette(tableau_color_blind)
plt.figure(figsize=(15,6))
ax = sns.pointplot(x="Hour", y="Counts", hue="WeekDay", data=summary, palette = current_palette,)
handles,labels = ax.get_legend_handles_labels()
#reordering legend content
handles = [handles[1], handles[5], handles[6], handles[4], handles[0], handles[2], handles[3]]
labels = [labels[1], labels[5], labels[6], labels[4], labels[0], labels[2], labels[3]]
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax.set_xlabel('Hour', fontsize = 12)
ax.set_ylabel('Count of Cabify Pickups', fontsize = 12)
ax.set_title('Hourly Cabify Pickups By Day of the Week in Peru (Jan-X 2010)', fontsize=16)
ax.tick_params(labelsize = 8)
ax.legend(handles,labels,loc=0, title="Legend", prop={'size':8})
ax.get_legend().get_title().set_fontsize('8')
plt.show()
```

```python
# Best idea is to copy a backup of original dataset and it transform
cabify2010bak = cabify2010
```

```python
cabify2010bak['start_at_ptz'] = cabify2010bak['start_at'].dt.tz_localize('Etc/GMT+1').dt.tz_convert('America/Lima')
```

```python
cabify2010bak['start_at_ptz'].head()
```

```python
cabify2010['start_at'].head()
```

Let's create new datetime again with peru timezone variables in our dataset (months, minutes, hours, day, weekday)

```python
#extracting the Hour from start_at column and putting it into a new column named 'Hour'
cabify2010bak['Hour'] = cabify2010bak.start_at_ptz.apply(lambda x: x.hour)

#extracting the Minute from start_at column and putting it into a new column named 'Minute'
cabify2010bak['Minute'] = cabify2010bak.start_at_ptz.apply(lambda x: x.minute)

#extracting the Month from start_at column and putting it into a new column named 'Month'
cabify2010bak['Month'] = cabify2010bak.start_at_ptz.apply(lambda x: x.month)

#extracting the Day from start_at column and putting it into a new column named 'Day'
cabify2010bak['Day'] = cabify2010bak.start_at_ptz.apply(lambda x: x.day)

#extracting the Weekday from start_at column and putting it into a new column named 'WeekDay'
cabify2010bak['WeekDay'] = cabify2010bak.start_at_ptz.apply(lambda x: x.strftime('%A'))
```

```python
#group the data by Weekday and hour
summary = cabify2010.groupby(['WeekDay', 'Hour'])['start_at_ptz'].count()
```

```python
#reset index
summary = summary.reset_index()
#convert to dataframe
summary = pd.DataFrame(summary)
#browse data
summary.head()
```

```python
#rename last column
summary=summary.rename(columns = {'start_at_ptz':'Counts'})
```

```python
sns.set_style('whitegrid')

## set palette   
current_palette = sns.color_palette(tableau_color_blind)
plt.figure(figsize=(15,6))
ax = sns.pointplot(x="Hour", y="Counts", hue="WeekDay", data=summary, palette = current_palette,)
handles,labels = ax.get_legend_handles_labels()
#reordering legend content
handles = [handles[1], handles[5], handles[6], handles[4], handles[0], handles[2], handles[3]]
labels = [labels[1], labels[5], labels[6], labels[4], labels[0], labels[2], labels[3]]
ax.yaxis.set_major_formatter(FormatStrFormatter('%.0f'))
ax.set_xlabel('Hour', fontsize = 12)
ax.set_ylabel('Count of Cabify Pickups', fontsize = 12)
ax.set_title('Hourly Cabify Pickups By Day of the Week in Peru (Jan-X 2010)', fontsize=16)
ax.tick_params(labelsize = 8)
ax.legend(handles,labels,loc=0, title="Legend", prop={'size':8})
ax.get_legend().get_title().set_fontsize('8')
plt.show()
```

Voila! Now it make sense, peak hour is between 8-10am, and after 5pm.
